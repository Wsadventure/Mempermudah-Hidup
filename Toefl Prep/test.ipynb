```python
for i in range(len(scoring_table)):
    temp_arr = scoring_table['Number-Right Score Range'].iloc[i].split('-')
    scoring_table['Number-Right Score Range'].iloc[i] = list(range(int(temp_arr[0]), int(temp_arr[1]) + 1))

scoring_table
```

```python
Section_1 = int(input('Enter the amount of questions you got right in Section 1'))
Section_2 = int(input('Enter the amount of questions you got right in Section 2'))
Section_3 = int(input('Enter the amount of questions you got right in Section 3'))
```

```python
input_array = [Section_1, Section_2, Section_3]
```

```python
scoring_table['Section 1 Converted Score'].values
```

```python
scoring_table['Section 1 Converted Score']
```

```python
for i in range(len(scoring_table)):
    if Section_1 in scoring_table['Number-Right Score Range'].iloc[i]:
        # Section_1_Score = 
        print(f'Section_1 of {Section_1} == index{i}')
```

```python
Section_Scores = []
for j in range(1,4):
    for i in range(len(scoring_table)):        
        if input_array[j-1] in scoring_table['Number-Right Score Range'].iloc[i]:
            Section_Scores.append(list(scoring_table[f'Section {j} Converted Score'].iloc[i].split('-')))
            print(f'Section {j} Converted Score is {Section_Scores[j-1]}')
print(Section_Scores)
```

```python
lower_bounds = 0
upper_bounds = 0
print('Scaled Score of your TOEFL ITP is')
for i in range(len(Section_Scores)):
    lower_bounds += int(Section_Scores[i][0])
    upper_bounds += int(Section_Scores[i][1])
    print(f'Section-{i} : {input_array[i]} -> {Section_Scores[i][0]}-{Section_Scores[i][1]}')

print(f'Total Scores : {lower_bounds} * 10/3 - {upper_bounds} * 10/3 -> {round(lower_bounds*10/3)}-{round(upper_bounds*10/3)}')
```
